```python
import pandas as pd
import numpy as np
```

```python
pc = pd.read_csv('Port_and_Commodity_TransBorder_-_January_2006_to_Latest_Available_Month.csv')
code = pd.read_csv('port_code.csv')
```

```python
pc
```

```python
code
```

```python
df3 = pd.merge(pc , code)
df3
```

```python
tra = pd.read_csv('transp.csv')
tra
```

```python
df4 = pd.merge(df3 , tra)
df4
```

```python
tcc = pd.read_csv('trade.csv')
tcc
```

```python
df5 = pd.merge(df4 , tcc)
df5
```

```python
ctry = pd.read_csv('country.csv')
ctry
```

```python
df6 = pd.merge(df5 , ctry)
df6
```

```python
Comm = pd.read_csv('Comm_code.csv')
Comm
```

```python
df7 = pd.merge(df6 , Comm)
df7
```

```python
df7['dateInt']=df7['Year'].astype(str) + df7['Month'].astype(str).str.zfill(2)
df7['Year_Month'] = pd.to_datetime(df7['dateInt'], format='%Y%m')
```

```python
df7
```

```python
df7.to_csv('2006_2021_US.csv',index=False)
```
